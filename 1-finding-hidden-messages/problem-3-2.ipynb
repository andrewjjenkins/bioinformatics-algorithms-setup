## Additional Exercise 3-2 experiments/follow-ups

```python
# This is not required for the exercise, just demonstrating pandas

from collections import Counter
[countsOrig, countsRc] = [Counter(s), Counter(rc)]
pd.DataFrame.from_dict({"Original": countsOrig, "RC": countsRc})
```

```python
# This is not required for the exercise, just demonstrating matplotlib

labels = ['C/G', 'G/C', 'T/A', 'A/T']
barsOrig = [countsOrig[x] for x in list('CGTA')]
barsRc = [countsRc[x] for x in list('GCAT')]
fig, ax = plt.subplots()
ax.bar(labels, barsOrig, 0.5, label='Original')
ax.bar(labels, barsRc, 0.5, bottom=barsOrig, label='Reverse Complement')
ax.set_ylabel('Count')
ax.set_title('Count of Nucleotides in original string and Reverse Complement')
ax.legend()
plt.show()
```
